```python
class Salary:
    def __init__(self,pay):
        self.pay=pay
    def get_total(self):
        return(self.pay*12)

class Employee:
    def __init__(self,pay,bonus):
        self.pay=pay
        self.bonus=bonus
        self.obj_salary=Salary(self.pay)
    def annual_sal(self):
        return "Annual salary is "+str(self.obj_salary.get_total())
print(Employee(365.7,100).annual_sal())
```

```python
class MyDate:
    def __init__(self,dd,mm,yy):
        self.dd=dd
        self.mm=mm
        self.yy=yy
    def __str__(self):
        return str(self.dd)+"-"+str(self.mm)+"-"+str(self.yy)
    def key(self):
        return self.yy*365+self.mm*30+self.dd

d=MyDate(15,8,1947)
print(d)

class MyEvent:
    def __init__(self,dd,mm,yy,detail):
        self.Date=MyDate(dd,mm,yy)
        self.detail=detail
    def __str__(self):
        return str(self.Date)+"=>"+self.detail
    def key(self):
        return self.Date.key()
e=MyEvent(15,8,1947,'Independence day')
print(e)
list=[MyEvent(15,8,1947,'Independence day'),MyEvent(15,8,1947,'Independence day'),MyEvent(15,8,1947,'Independence day')]
for f in sorted(list,key=MyEvent.key):
    print(f)
```
